```python
import requests
import pandas as pd
import time
import urllib.parse

url = "https://api.github.com/search/repositories"
```

```python
languages = ["shell","lua","objective-c","powershell","c","r","ruby","julia","scala","groovy","swift","dart","starlark","webassembly","matlab","rust","elixir","erlang","python","java","javascript","nix","php","go","kotlin","clojure","ocaml","haskell","perl","typescript","vue","C%23","F%23","C%2B%2B"]
#languages = ["c","r"]
language = []
total_count = []
for i in languages:
    time.sleep(5)
    params = 'q=language:%s' % i
    print(params)
    res = requests.get(url=url, params=params)
    tc = res.json()
    language.append(urllib.parse.unquote(i))
    total_count.append(tc['total_count'])
```

```python
df = pd.DataFrame(
        {
            "Language": language,
            "Total Count": total_count,
        },
        index=language
    ).sort_values("Total Count", ascending=False)
```

```python
df
```

```python
plot = df.plot.pie(y='Total Count',figsize=(5, 5), legend=False)
plot.set_ylabel('Total Count', rotation=90, labelpad=100)
```
